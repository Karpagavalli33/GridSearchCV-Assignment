```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
indep
```

```python
dep
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid={'criterion' :['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
           'max_features':['auto','sqrt','log2'],'n_estimators':[10,100]}
```

```python
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit = True,verbose= 3,n_jobs=-1)
grid.fit(indep,dep)
```

```python
re=grid.cv_results_
print("The R Score Value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1"))
smoker_yes_input=int(input("Smoker yes 0 or 1"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future Prediction={}".format(Future_Prediction))
```
